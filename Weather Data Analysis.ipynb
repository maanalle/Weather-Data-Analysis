```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data['Date/Time'] = pd.to_datetime(data['Date/Time']) #Date/Time column is currently an object type, we should convert it to a datetime type for better handling
```

```python
data.dtypes
```

```python
#Temperature Over Time
plt.figure(figsize=(10, 6))
plt.plot(data['Date/Time'], data['Temp_C'], marker='o', linestyle='-', color='b')
plt.title('Temperature Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Wind Speed Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data['Wind Speed_km/h'], kde=True, color='green')
plt.title('Wind Speed Distribution')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Humidity vs Temperature
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Temp_C'], y=data['Rel Hum_%'], color='orange')
plt.title('Humidity vs Temperature')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
#Weather Condition Count
plt.figure(figsize=(8, 6))
sns.countplot(y=data['Weather Condition'], palette='Set2')
plt.title('Weather Condition Count')
plt.xlabel('Count')
plt.ylabel('Weather Condition')
plt.show()
```

```python
#Temperature and Dew Point Temperature Over Time
plt.figure(figsize=(12, 6))
plt.plot(data['Date/Time'], data['Temp_C'], label='Temperature (Â°C)', color='blue')
plt.plot(data['Date/Time'], data['Dew Point Temp_C'], label='Dew Point (Â°C)', color='red')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (Â°C)')
plt.title('Temperature and Dew Point Temperature Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Correlation Heatmap
# Calculate correlation matrix
corr = data[['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Visibility_km', 'Press_kPa']].corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Wind Speed vs. Temperature Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Temp_C'], y=data['Wind Speed_km/h'], color='purple')
plt.title('Wind Speed vs. Temperature')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Wind Speed (km/h)')
plt.show()
```

```python
#Monthly Temperature Variation
# Extract month and year
data['Month'] = data['Date/Time'].dt.month
data['Year'] = data['Date/Time'].dt.year

# Group by month and year, and calculate the average temperature
monthly_temp = data.groupby(['Year', 'Month'])['Temp_C'].mean().unstack()

# Plot the data
plt.figure(figsize=(12, 6))
monthly_temp.plot(kind='line', marker='o')
plt.title('Average Monthly Temperature')
plt.xlabel('Year')
plt.ylabel('Average Temperature (Â°C)')
plt.legend(title='Month', labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Weather Condition Frequency by Year
data['Year'] = data['Date/Time'].dt.year

plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Year', hue='Weather Condition', palette='Set1')
plt.title('Weather Condition Frequency by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Pressure vs. Temperature
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Temp_C'], y=data['Press_kPa'], color='magenta')
plt.title('Pressure vs. Temperature')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Pressure (kPa)')
plt.show()
```
